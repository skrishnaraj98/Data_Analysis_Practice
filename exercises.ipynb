```python
ax = Sales['Order_Quantity'].plot(kind = 'box', vert = False, figsize = (10,3))
plt.xticks(range(0,35))

Q1 = Sales['Order_Quantity'].quantile(0.25)
Q2 = Sales['Order_Quantity'].quantile(0.50)
Q3 = Sales['Order_Quantity'].quantile(0.75)

plt.axvline(Q1, color = 'r', linestyle = ':', label = '1st Quartile (Q1)')
plt.axvline(Q2, color = 'g', linestyle = 'dashdot', label = '2nd Quartile (Median)')
plt.axvline(Q3, color = 'b', linestyle = 'solid', label = '3rd Quartile (Q3)')
```

```python
ax = Sales['Order_Quantity'].plot(kind = 'box', vert = False, figsize = (12,2))

plt.xticks(range(0,41))

Q1 = Sales['Order_Quantity'].quantile(0.25)
Q2 = Sales['Order_Quantity'].quantile(0.50)
Q3 = Sales['Order_Quantity'].quantile(0.75)

plt.axvline(Q1, color = 'r', linestyle = 'dashdot', label = '1st Quartile')
plt.axvline(Q2, color = 'g', linestyle = 'dashdot', label = '2nd Quartile')
plt.axvline(Q3, color = 'b', linestyle = 'dashdot', label = '3rd Quartile')

plt.annotate('1st Quartile', xy=(Q1, 1), xytext=(Q1, 1.2),
             arrowprops=dict(facecolor='r', arrowstyle='->'),
             horizontalalignment='left', color='r')
plt.annotate('2nd Quartile', xy=(Q2, 1), xytext=(Q2, 1.2),
             arrowprops=dict(facecolor='g', arrowstyle='->'),
             horizontalalignment='left', color='g')
plt.annotate('3rd Quartile', xy=(Q3, 1), xytext=(Q3, 1.2),
             arrowprops=dict(facecolor='b', arrowstyle='->'),
             horizontalalignment='left', color='b')
```
