```python
import numpy as np
import pandas as pd
```

```python
delivery = pd.read_csv("Downloads//data//deliveries.csv")
```

```python
matches = pd.read_csv("Downloads//data//matches.csv")
```

```python
delivery.head(2)
```

```python
matches.head(2)
```

```python
## Now we are going to merge these 2 datasets through a common column "id" which acts as the linking bridge
## to find out the top scoring batsman in every season. 
```

```python
delivery.merge(matches, left_on = "match_id", right_on = "id")
```

```python
new = delivery.merge(matches, left_on = "match_id", right_on = "id")
```

```python
new.shape
```

```python
new.groupby(["season","batsman"])["batsman_runs"].sum()
```

```python
new.groupby(["season","batsman"])["batsman_runs"].sum().sort_values(ascending = False).reset_index()
```

```python
final = new.groupby(["season","batsman"])["batsman_runs"].sum().sort_values(ascending = False).reset_index()
```

```python
final.drop_duplicates(subset ="season", keep="first").sort_values("season")
```
